#6. Anda dapat menggunakan model pipeline untuk mempermudah perkejaan Anda.



```python
pipe_filter = Pipeline([
    ("std", StandardScaler()),
    ("selector", SelectKBest(f_classif, k= 20)),
    ("clf", LogisticRegression(max_iter=1000))
])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)
pipe_filter.fit(X_train, y_train)
pred = pipe_filter.predict(X_test)
print("=== Filter (ANOVA) + LR ===")
print("Accuracy:", accuracy_score(y_test, pred))
print(classification_report(y_test, pred))
```

```python
# ambil nama asli kecuali kolom yang telah didrop
original_features = df.drop(columns=['diagnosis']).columns
# Get the indices of the selected features
selected_indices = pipe_filter.named_steps["selector"].get_support(indices=True)
# Get the names of the selected features using the original feature names and the indices
feat_names = original_features[selected_indices]
print("Nama Fitur: ", feat_names.tolist())
```
